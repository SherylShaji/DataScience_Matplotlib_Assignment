### Data Cleaning 

```python
%matplotlib inline
```

```python
# Import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Read the mouse data and the study results
csv_path1 = os.path.join('../','Resources', 'Mouse_metadata.csv')
mousedata_df = pd.read_csv(csv_path1)

csv_path2 = os.path.join('../','Resources', 'Study_results.csv')
studyresults_df = pd.read_csv(csv_path2)
```

```python
# Display the mouse data
mousedata_df
```

```python
# Display the study data
studyresults_df
```

```python
# Combine the data into a single dataset and display it
combined_df = pd.merge(mousedata_df,studyresults_df,on='Mouse ID',how='outer')
combined_df
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates_df = combined_df[combined_df.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_mouseid_df=duplicates_df['Mouse ID'].unique()
duplicate_mouseid_df
```

```python
# Optional: Get all the data for the duplicate mouse ID.

duplicate_mouseid_data_df=combined_df.loc[combined_df['Mouse ID'] == 'g989']
duplicate_mouseid_data_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID and display it
combinedcopy_df=combined_df.copy()

clean_df=combinedcopy_df.loc[combinedcopy_df['Mouse ID'] != 'g989']
clean_df
```

### Generating the Boxplot

```python
# Determine the final timepoint for each mouse.
# Start by getting the greatest timepoint for each mouse
final_timepoint_df=clean_df[['Mouse ID','Timepoint']]
greatest_df=final_timepoint_df.groupby('Mouse ID')[['Timepoint']].max()
greatest_df
greatest_df.reset_index()
```

```python
# Join the newly created `.max()` dataframe to the dataframe from Part 1
updated_df = pd.merge(clean_df,greatest_df,on=['Mouse ID','Timepoint'],how='right')
updated_df
```

```python
# Create a list with all 10 drug regimens.
drug_regimes=['Ceftamin','Capomulin','Infubinol','Ramicane','Zoniferol','Propriva','Placebo','Ketapril','Naftisol','Stelasyn']
# Create a empty list to fill with the tumor vol data
tumor_vol=[]

# Isolate (filter) each mice on each drug to collect their tumor volume.
for i in drug_regimes:
# Locate the rows which match the drug and get the final tumor volumes of all mice
# Append the outcome to the empty list previously created.
    tumor_vol.append(updated_df.loc[updated_df['Drug Regimen'] == i]['Tumor Volume (mm3)'])
```

```python
# Create a boxplot that visualizes the final tumor volume of all mice in the study across all drug regimens.
# Define a custom shape for all outliers in the visualization
# Create horizontal box and whisker plot

box=plt.boxplot(tumor_vol,vert=False,sym='g*')
for m in box['medians']: 
    m.set(color ='red', 
               linewidth = 2) 
for c in box['caps']: 
    c.set(color ='blue', 
            linewidth = 2)
for w in box['whiskers']: 
    w.set(color ='green', 
                linewidth = 2)
plt.xlabel('Final tumor volume (mm3)')
plt.ylabel('Drug Regimes')
plt.title('Boxplot Visualization')
plt.yticks(np.arange(1, 11, 1),drug_regimes)
plt.show()
```
